```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_org = pd.read_csv('E:\DATA SCIENCE\supermarket_sales - Sheet1.csv')
df_org
```

```python
df_org.columns
```

```python
df_org = df_org.drop(['Invoice ID','Unit price','Quantity','Tax 5%','Time'],axis=1)
df_org.columns
```

```python
cols_list = df_org.columns
cols_list = cols_list[-4:]
df_org = df_org.drop(cols_list,axis=1)
df_org
```

# Top 3 Branches With Sales

```python
df_branch = df_org[['Branch','Total']]
df_branch = df_branch.groupby('Branch').sum().reset_index()
df_branch.sort_values(by='Total', ascending=False, na_position='last',inplace=True)
branch_data_top_3 = {}
for i in range(0,3):
    branch_data_top_3[df_branch.iloc[i][0]] = df_branch.iloc[i][1]

df_branch
```

# Overall Categories Sales

```python
df_cat = df_org[['Product line','Total']]
df_cat = df_cat.groupby('Product line').sum().reset_index()
df_cat.sort_values(by='Total',inplace=True,ascending=False,na_position='last')
categories_data_top_3={}
categories_data_bottom_3={}
for i in range(0,6):
    if(i<=2):
        categories_data_top_3[df_cat.iloc[i][0]] = df_cat.iloc[i][1]
    else:
        categories_data_bottom_3[df_cat.iloc[i][0]] = df_cat.iloc[i][1]

df_cat
```

# Members & Non Members Stats 

```python
df_customer = df_org[['Customer type','Total']]
df_customer = df_customer.groupby('Customer type').sum().reset_index()
df_customer.sort_values(by='Total',inplace=True,ascending=False,na_position='last')
customer_type_data = {}
for i in range(0,2):
    customer_type_data[df_customer.iloc[i][0]] = df_customer.iloc[i][1]

df_customer
```

# Male & Female Stats 


```python
df_gender = df_org[['Gender','Total']]
df_gender = df_gender.groupby('Gender').sum().reset_index()
df_gender.sort_values(by='Total',inplace=True,ascending=False,na_position='last')
gender_sales = {}
for i in range(0,2):
    gender_sales[df_gender.iloc[i][0]] = df_gender.iloc[i][1]

gender_sales
```

```python
# Categories with Male Ratio
df_male_items = df_org[df_org['Gender']=='Male']
df_male_items = df_male_items[['Product line','Total']]
df_male_items = df_male_items.groupby('Product line').sum()
df_male_items = df_male_items.reset_index()
df_male_items.sort_values(by='Total',ascending=False,inplace=True,na_position='last')


males_items_stats = {}
for i in range(0,6):
    males_items_stats[df_male_items.iloc[i][0]] = df_male_items.iloc[i][1]

males_items_stats
```

```python
# Categories with Female Ratio
df_female_items = df_org[df_org['Gender']=='Female']
df_female_items = df_female_items[['Product line','Total']]
df_female_items = df_female_items.groupby('Product line').sum()
df_female_items = df_female_items.reset_index()
df_female_items.sort_values(by='Total',ascending=False,inplace=True,na_position='last')

females_items_stats = {}
for i in range(0,6):
    females_items_stats[df_female_items.iloc[i][0]] = df_female_items.iloc[i][1]

females_items_stats
```

# Payment Method

```python
df_payment = df_org[['Payment','Total']]
df_payment = df_payment[['Payment','Total']]
df_payment = df_payment.groupby('Payment').sum().reset_index()
df_payment.sort_values(by='Total',ascending=False,inplace=True,na_position='last')
payment_data= {}
for i in range(0,3):
    payment_data[df_payment.iloc[i][0]] = df_payment.iloc[i][1]

payment_data
```
